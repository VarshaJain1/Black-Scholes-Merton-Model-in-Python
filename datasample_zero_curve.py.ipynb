```python
# data/sample_zero_curve.py

import pandas as pd
import matplotlib.pyplot as plt

def load_sample_zero_curve():
    """
    Returns a sample zero-coupon yield curve as a Pandas DataFrame.
    Columns: ['Maturity (Years)', 'Zero Rate']
    """
    data = {
        "Maturity (Years)": [0.5, 1, 1.5, 2, 2.5, 3, 3.5],
        "Zero Rate": [0.050, 0.061, 0.072, 0.075, 0.080, 0.085, 0.090]
    }
    df = pd.DataFrame(data)
    return df

def plot_zero_curve(df):
    """
    Plots the zero-coupon yield curve.
    """
    plt.figure(figsize=(8,5))
    plt.plot(df["Maturity (Years)"], df["Zero Rate"], marker="o", linestyle="-", color="blue")
    plt.title("Sample Zero-Coupon Yield Curve")
    plt.xlabel("Maturity (Years)")
    plt.ylabel("Zero Rate")
    plt.grid(True, alpha=0.3)
    plt.show()

# -----------------------
# Example Run
# -----------------------
if __name__ == "__main__":
    curve = load_sample_zero_curve()
    print(curve)
    plot_zero_curve(curve)
```
